**HYPOTHESIS TESTING**

**Income across marital status, education and dt_cutomer**

```python
#ANOVA for MARITAL STATUS
model = ols('Income ~ C(Q("Marital_Status"))', data=mkt_data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Results for Marital statuses:")
print(anova_table)
# Check for significance
if anova_table['PR(>F)'][0] < 0.05:
    print("\nSignificant differences found among marital status.\n")
else:
    print("\nNo significant differences among marital status.\n")
```

```python
#ANOVA for EDUCATION
model = ols('Income ~ C(Q("Education"))', data=mkt_data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Results for Education:")
print(anova_table)
# Check for significance
if anova_table['PR(>F)'][0] < 0.05:
    print("\nSignificant differences found among education.\n")
else:
    print("\nNo significant differences among education.\n")
```

```python
#ANOVA for DT_customer
model = ols('Income ~ C(Q("Dt_Customer"))', data=mkt_data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Results for customer join dates:")
print(anova_table)
# Check for significance
if anova_table['PR(>F)'][0] < 0.05:
    print("\nSignificant differences found among customer join dates.\n")
else:
    print("\nNo significant differences among customer join dates.\n")
```
